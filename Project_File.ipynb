### 1) **Data Desciption**

#### **Collection Method:**

Data was collected from a study done by the Pacific Laboratory for Artifical Intelligence (PLAI), a computer science research group led by Dr. Frank Wood. Their current project involves the creation an embodied AI model, which is an AI able to recieve, understand, and respond to a complex environment, similar to how humans interact with the world around them.

To do so, the popular game: "Minecraft" was chosen as the "complex environment". Behavioural data, such as locations visited, what players said in chat, in game actions, etc. were taken from players of the "PLAICRAFT" server, a free Minecraft server ran by PLAI, in order to study the trends and patterns of how players interact with the virtual world. This data will then be used to train and develop the AI model.

Information from the players.csv dataset was retreived primarily from the signup process of each player as well as some tracking of their play. On the contrary, all of the data besides the hashed E-mail used to identify the player, was retreived purely from player gameplay within the sessions.csv dataset.

* **Potential Issues:** <br>

Within the sessions.csv dataset there are a few issues which are needed to be taken into account. The first one is that the start_time and end_time contain both the time of the sessions as well as the date of the session, this is untidy and will need to be split. Secondly, the datatype of the `start_time` and `end_time` variables is character (chr) which will be needed to be converted to double (dbl) or integer (int) in order to perform calculations (e.g average play time in hours on x day). Lastly, the `original_start_time` and `original_end_time` are in a very starnge format which is hard to understand, conversion into a more drect format would benefit clarity of the data.
