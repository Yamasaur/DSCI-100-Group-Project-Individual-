#### **Collection Method:**

Data was collected from a study done by the Pacific Laboratory for Artificial Intelligence (PLAI), a computer science research group led by Dr. Frank Wood. Their current project involves the creation of an embodied AI model, which is an AI able to receive, understand, and respond to a complex environment, similar to how humans interact with the world around them.

To do so, the popular game: "Minecraft" was chosen as the "complex environment". Behavioural data, such as locations visited, what players said in chat, in game actions, etc. were taken from players of the "PLAIcraft" server, a free Minecraft server run by PLAI, in order to study the trends and patterns of how players interact with the virtual world. This data will then be used to train and develop the AI model.

Information from the players.csv dataset was retrieved primarily during the signup process of each player with the rest coming from their actual play in the server. On the contrary, all of the data on the sessions.csv dataset was retrieved purely from player gameplay besides the hashed E-mail used to identify the player.

### 4) **Methods and Plan**

The question posed in part 2) goes as follows:

"Can the number of hours a player plays and their age predict their skill/experience level?"

In order to provide a meaningful answer, I believe a classification method would be most appropriate.

**Why is this method appropriate?**

This method would be appropriate as it is a predictive type of question. This form of question asks for an expected outcome such as a value or label, based on prior data regarding the subject known as predictors. An example of this would be to predict whether a type of cancer is malignant or benign based upon its characteristics like size. In regards to question 2, the label we want to predict would be the experience level and the predictors would be the player age and total hours they played. In order to answer predictive problems, classification is used.

**Which assumptions are required, if any, to apply the method selected?**

The key assumption is that the data is evenly distributed, the reason being that classification relies on the similarity between points, thus to achieve accurate results, trends within the grouping/trends of data points is required.

**What are the potential limitations or weaknesses of the method selected?**

The main limitation of classification is the need for high intensity and relatively even distribution. This is because datapoints which are considered outliers can play a heavy effect on the final prediction of a new point. For example, if a new player enters their age as 60, classification of them would be based on all other players who are around 60. What happens if there is only one previous 60 year old player and the rest of the dataset contains players 10 years old or under? The model will then base its prediction off of this very limited data making that prediction’s accuracy very low.

**How are you going to compare and select the model?**

So far, we have learned knn and linear regression as classification techniques/models. Linear regression is used to predict values from a known linear relationship, meanwhile knn is used to predict values with a non-linear relationship. In this case the experience level (Amateur, Beginner, Regular, Veteran, Pro), age, and hours played all have non-linear relationships with one another. Thus a knn model will be the most appropriate.

**How are you going to process the data to apply the model? For example: Are you splitting the data? How? How many splits? What proportions will you use for the splits? At what stage will you split? Will there be a validation set? Will you use cross validation?**

Data will be first split into training and testing sets with a 70-30 split as the dataset only contains 195 observations. The model will be made or trained using the training sets. Within the training sets, cross validation will be used to determine the best k value. From our visualisations, we can see that datapoints are pretty intense on specific areas thus a smaller k value will suffice as to not overfit the data. To determine the actual best k-value of the dataset with 3-6 validations sets being used to represent each potential k value. From there a knn model will be made using `Age` and `played_hours` as predictors to predict `experience`. When designing the model, `NA` values in the `Age` predictor and values equal to 0 in the `played_hours` predictor will be removed as they do not provide any meaningful data for our classification and may affect its overall performance.
